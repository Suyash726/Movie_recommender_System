```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
%matplotlib inline
```

```python
df = pd.read_csv('movie_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# The most important features for our project are genres,cast,director,keywords.So we will start working on those 4 features first
All_features = ["keywords",'genres','cast','director']

df[All_features].head()
```

```python
# Replacing the empty values in those 4 columns witn ''
for feature in All_features:
    df[All_features] = df[All_features].fillna('')
```

```python
df.head()
```

```python
# Function which returns a combination of all four features when a row is passed
def Combined_features(row):
    return row['keywords'] +" "+row['cast']+" "+row["genres"]+" "+row["director"]
```

```python
# Lets create a column containig all the required functions
df['Combined_features'] = df.apply(Combined_features,axis=1)
```

```python
# Here we are using the countvectorizer method to convert the collection of text elements to a matrix of token elements
vectorizer = CountVectorizer()
Feat_vector = vectorizer.fit_transform(df['Combined_features'])
# The cos_similarity method returns the similarity index between two specified objects
cos_sim = cosine_similarity(Feat_vector)
print(cos_sim)
```

```python
def get_title(index):
    return df[df.index == index]['original_title'].values[0]
```

```python
def get_index(title):
    title = title
    return df[df.title == title]["index"].values[0]
```

```python
Movie_i_like = 'Tropic Thunder'
```

```python
movie_index = get_index(Movie_i_like)
```

```python
similar = list(enumerate(cos_sim[movie_index]))

sorted_movies = sorted(similar,key=lambda x:x[1],reverse=True)
```

```python
print (sorted_movies)
```

```python
print("Top 10 Similar Movies to "+ Movie_i_like + " are :")
i = 0
for i in range(len(sorted_movies)):
    print("Movie :" + get_title(sorted_movies[i][0]),",Similarity Score : " , sorted_movies[i][1])
    print()
    i=i+1
    if i > 10:
        break
```
